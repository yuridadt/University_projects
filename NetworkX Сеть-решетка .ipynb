```
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

nx.__version__
```

## Программа строит  граф-решетку



Зададим N- количество решеток, K -кол-во нодов в ней

```
K=4
N=3
```

```
G = nx.grid_2d_graph(N*K-(N-1), N*K-(N-1))
nx.draw(G)
```

```
K=5
N=7

#задаем граф
G=nx.grid_2d_graph(N*K-(N-1), N*K-(N-1))

#задаем структуру решетки
pos = dict( (n,n) for n in G.nodes() )
#задаем цифры ( для удобства)
labels = dict( ((i, j), i * 10 + j) for i, j in G.nodes() )


sdvig_y=1
#отрисовка столбца
for y in range(N):
  sdvig_x=1
  #отрисовка строки
  for x in range(N):
    #отрисовка квадратика
    for i in range(K-2):
      for k in range(K-2):
        G.remove_node((sdvig_x+i,sdvig_y + k))
        labels.update({(sdvig_x+i,sdvig_y + k): ''})
  
    sdvig_x+=(K-1)
  sdvig_y+=(K-1)
#рисуем график
plt.figure(figsize=(10,10))
nx.draw(G, pos=pos, labels=labels)
```
